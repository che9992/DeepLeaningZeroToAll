### Problem with Linear regression 
- #### We know Y is 0 or 1 
- #### Hypothesis can give values large than 1 or lees than 0 
<hr/>

### Logistic Hypothesis
- #### \\(g(z) = \dfrac{1}{1+e^{-z}}\\)
- #### \\(H(X) = \dfrac{1}{1+e{-W^TX}}\\)

### New cost function for logistic
- #### \\( cost(W) = \frac{1}{m}\sum\\)    
- #### \\(   c(H(x),y)\\)
- #### \\(c(H(x),y) = {-log(H(x)) : y = 1} ,{-log(1-H(x)) : y = 0}\\)
- #### with out if condition
 
 - #### \\(c(H(x),y) = -ylog(H(x))-(1-y)log(1-H(x))\\)

### Minimize cost - Grandient decent algorithm
- #### \\(cost(W) = -\frac{1}{m}\sum{ylog}(H(x))+(1-y)log(1-H(x))\\)
- #### \\(W:= W- \alpha\dfrac{\alpha}{\alpha{W}}cost(W)\\)

```python
# cost function 
cost = -tf.reduce_mean((Y * tf.log(hypothesis) + (1 - Y) * tf.log(1 - hypothesis)))
# minimize
a = tf.Variable(0.1) #Learning rate, alpha
optimizer = tf.train.GrandientDescentOptimizer(a)
train = optimizer.minimize(cost)
```

## example

````python
import tensorflow as tf

x_data = [ [1, 2], [2, 3], [3, 1], [4, 3], [5, 3], [6, 2] ]
y_data = [ [0], [0], [0], [1], [1], [1] ] # 0 is fail , 1 is pass  

X = tf.placeholder(tf.float32, shape = [None, 2])
Y = tf.placeholder(tf.float32, shape = [None, 1])
W = tf.Variable(tf.random_normal([2, 1]), name = 'weight')
b = tf.Variable(tf.random_normal([1]), name = 'bias' )

hypothesis = tf.sigmoid(tf.matmul(X, W) + b)
cost = -tf.reduce_mean(Y * tf.log(hypothesis) + (1 - Y) 
                                      * tf.log(1 - hypothesis))

train = tf.train.GradientDescentOptimizer(learning_rate=0.01).minimize(cost)

# Accuracy computation
# True if hypothesis > 0.5 else False 
# tf.cast makes it to 0 or 1 
predicted = tf.cast(hypothesis > 0.5, dtype=tf.float32)
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, Y), dtype=tf.float32))


with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    for step in range(10001):
        cost_val, _ = sess.run([cost, train], feed_dict={X: x_data, Y: y_data})
        if step % 200 == 0:
            print(step, cost_val)
    
    h, c, a = sess.run([hypothesis, predicted, accuracy],
                      feed_dict={X: x_data, Y: y_data})
    print("\nHypothesis:\n",h,"\nCorrect (Y):\n",c,"\nAccuracy:\n",a)
````

## example with data file
### Classifying diabetes

````python
import numpy as np 

xy = np.loadtxt('data-03-diabetes.csv', delimiter = ',', dtype = np.float32)
x_data = xy[:, 0:-1]
y_data = xy[:, [-1]]

X = tf.placeholder(tf.float32, shape=[None, 8])
Y = tf.placeholder(tf.float32, shape=[None, 1])
W = tf.Variable(tf.random_normal([8, 1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name = 'bias')

hypothesis = tf.sigmoid(tf.matmul(X, W) + b)
cost = -tf.reduce_mean(Y * tf.log(hypothesis) + (1 - Y) * tf.log(1 - hypothesis))
train = tf.train.GradientDescentOptimizer(learning_rate=1e-2).minimize(cost)

predicted = tf.cast(hypothesis > 0.5, dtype=tf.float32)
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, Y), dtype=tf.float32))

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    feed = {X: x_data, Y: y_data}
    for step in range(10001):
        sess.run(train, feed_dict=feed)
        if step % 200 == 0:
            print(step, sess.run(cost, feed_dict=feed))
            
    h, c, a = sess.run([hypothesis, predicted, accuracy], feed_dict=feed)
    print('\nHypothesis:\n', h, '\nCorrect:\n', c, '\nAccuracy:\n', a * 100,'%')
    
````
